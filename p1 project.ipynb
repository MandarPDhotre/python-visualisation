```python
n = 20000
mean = [0, 0]
cov = [(2, .4), (.4, .2)]
rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(mean, cov, n).T

f, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(x=x, y=y, levels=5, color="w", linewidths=1)

sns.set_theme(style='dark')
```

```python
pd.plotting.scatter_matrix(df, alpha=0.5, diagonal='hist')
plt.title('Scatter Matrix')
plt.show()
```

```python
plt.fill_between(df['foundingyear'], df['price'], alpha=0.5) #Area chart
plt.plot(df['foundingyear'], df['price'])
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Area chart')
plt.show()
```

```python
plt.stem(df['society'], df['price'], markerfmt='ro', linefmt='r--', basefmt='k-')
plt.xticks(df['society'])
plt.xlabel('Society')
plt.ylabel('Price')
plt.title('lollipop chart')

plt.show()
```

```python
from matplotlib_venn import venn2
```

```python
data1 = {'A': [1, 2, 3, 4],
         'B': [3, 4, 5, 6]}

data2 = {'A': [3, 4, 5, 6],
         'B': [1, 2, 3, 4]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

fig, ax = plt.subplots(figsize=(8, 6))

venn2([set(df1['A']), set(df2['A'])], set_labels=['Group 1', 'Group 2'], ax=ax)

ax.set_title('Venn Diagram')

plt.show()
```
